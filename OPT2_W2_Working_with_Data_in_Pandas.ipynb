### Using `.iloc` to Access DataFrame Elements

The `.iloc` indexer is primarily integer-location based (from 0 to `length-1` of the axis). It can take a single integer, a list of integers, or a slice object for rows and columns. It allows you to select:

*   **A specific row:** `df.iloc[row_index]`
*   **A specific column:** `df.iloc[:, col_index]` (note the colon for all rows)
*   **A specific cell:** `df.iloc[row_index, col_index]`
*   **Ranges of rows and columns:** `df.iloc[start_row:end_row, start_col:end_col]`

````python
# Display the first 5 rows to remember the structure (iloc uses 0-based indexing)
display(income.head())
````

#### Accessing the first row using `.iloc`

````python
# Access the first row (index 0) of the DataFrame
display(income.iloc[0])
````

#### Accessing the first column using `.iloc`

````python
# Access the first column (index 0) of the DataFrame (all rows, first column)
display(income.iloc[:, 0].head())
````

#### Accessing a specific cell using `.iloc` (e.g., value at row 0, column 0)

````python
# Access the value in the first row (index 0) and first column (index 0)
display(income.iloc[0, 0])
````

#### Accessing a range of rows and columns (e.g., rows 1-5, columns 0-2)

````python
# Access rows from index 1 up to (but not including) 6, and columns from index 0 up to (but not including) 3
display(income.iloc[1:6, 0:3])
````

## Example: Cleaning data

````python
clean = income[['Childless couple, annual income','Couple, two children under 14']]
clean.head()
````

We see those pesky commas. Now we can get on with cleaning up the data:

````python
clean=clean.replace(',', '', regex=True)

# In addition, missing values are sometimes written as '-', in order for Python to understand that it is just a missing numerical
# value, all '-' need to be replaced with 'NaN'.
clean = clean.replace('-', 'NaN', regex=True).astype('float')
clean.head()
````

**Extension**: "**Regex**" refers to "**Reg**ular **Ex**pression", which is a way of replacing and cleaning text. It's a bit beyond the scope of this class, but worth looking into if you're interested in programming more widely.

This seems to have done the job. We've also put a line in the code to get rid of dashes - a way that data collectors will sometimes represent missing data. Now let's plot this.

# Extension: Web Scraping

In this example, we've been working with a .csv file that contains all the data we want. That's not always the case. Let's say we're interested in getting the data from a table on a website. Websites are built using HTML code, so what we need to figure out how to look inside the website's code and pull out the data we want. Luckily, pandas has a built in function that can automatically recognize HTML tables in websites and turn them into dataframes.

Let's start with the [Netflix Top 10](https://top10.netflix.com/) website. Click on the link and have a look around. You'll notice two tables: the first showing the top 10 films this week, and the second (farther down) showing the most popular filsms based on their first 28 days on netflix.

We can download both of these tables into python using one pandas function: read_html

````python
url='https://top10.netflix.com/'

tables=pandas.read_html(url)

print(tables)
````

When we print the results of what was scraped, it's pretty ugly. One of the reasons is that the `tables` variable is actually a *list* of dataframes. Because there were two tables on our website, `read_html` has returned both of those tables and put them in a list. let's save the first table as a new dataframe called `top10` and have a closer look.

````python
top10=tables[0]
top10
````

This looks more like the dataframes we were looking at earlier. There's a big chunk of text (this is HTML code, the language websites are built with) where the name of the second column should be. `read_html` is usually pretty smart, and can actually read the column names from the tables on the website. It seems to have gotten confused for this one column. If we print the columns from the  We can rename that column using the `rename` function. Since we know it's the second column, we can select it with `top10.columns[1]`

````python
top10.rename(columns={top10.columns[1]: "Title" }, inplace = True)
top10
````

And there we have it; a nicely formatted dataframe ready for analysis, straight from a website.

#### Load Data from Wikipedia
Reasoning: Wikipedia blocks automated requests that lack proper browser identification, resulting in an HTTP Error 403: Forbidden. To bypass this restriction, I'll add a User-Agent header via the storage_options parameter in pandas.read_html(). This header mimics a standard web browser request, allowing successful data extraction from the page.

````python
# Assessed Question Solution
# Calculate average road fatalities per 100,000 inhabitants in Africa (2021)

import pandas as pd
import requests
import io

# URL for the Wikipedia page containing traffic death rate data
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_traffic-related_death_rate'

# User-Agent header to mimic a web browser and avoid HTTP 403 errors
headers = {'User-Agent': 'Mozilla/5.0'}

# Fetch page content and read HTML tables into a list of DataFrames
response = requests.get(url, headers=headers)
tables = pd.read_html(io.StringIO(response.text))

# The first table usually contains the country-level data we need
df = tables[0]

# Filter for African countries and 2021 data
# .copy() ensures we're working with an independent DataFrame, preventing SettingWithCopyWarning
africa_2021 = df[(df['Continent'] == 'Africa') & (df['Date[b]'] == '2021')].copy()

# Column name for per capita fatalities
fatalities_column = 'Per 100,000 inhabitants[7][1][8]'

# Convert the fatalities column to numeric, coercing errors to NaN
# This handles non-numeric characters like commas or hyphens, similar to previous cleaning tasks
africa_2021[fatalities_column] = pd.to_numeric(africa_2021[fatalities_column], errors='coerce')

# Calculate the average fatalities for Africa using groupby
result = africa_2021.groupby('Continent')[fatalities_column].mean()

# Display the final answer
print(f"Average road fatalities per 100,000 inhabitants in Africa (2021): {result['Africa']:.2f}")

# Display the list of African countries included in the calculation and their rates
print(f"\nNumber of African countries included: {len(africa_2021)}")
print("\nAfrican countries and their 2021 rates:")
print(africa_2021[['Country/region', fatalities_column]].sort_values(by=fatalities_column, ascending=False))
````
